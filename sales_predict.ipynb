```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
data = {
    'year': [2015, 2016, 2017, 2018, 2019, 2020],
    'sales': [10000, 12000, 14000, 16000, 18000, 20000]
}

df = pd.DataFrame(data)

# Training the linear regression model
X = df[['year']]
y = df['sales']

model = LinearRegression()
model.fit(X, y)

# Making predictions for future years
future_years = np.array([[2021], [2022], [2023]])
predictions = model.predict(future_years)

# Plotting the data and predictions
plt.scatter(df['year'], df['sales'], color='blue', label='Actual Sales')
plt.plot(np.concatenate([df['year'], future_years.flatten()]),
         np.concatenate([y, predictions]), color='red', label='Predicted Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Forecast')
plt.legend()
plt.show()
```
